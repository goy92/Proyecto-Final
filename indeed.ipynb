```python
import requests
import time
from bs4 import BeautifulSoup
from dotenv import load_dotenv
load_dotenv()
import os
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd
```

```python
def get_info_page():
    
    busqueda = "Consultor/a general"

    url = "https://www.indeed.es/"
    url_salarios = "https://www.indeed.es/salaries?from=headercmplink&attributionid=HOMEPAGE"

    driver = webdriver.Firefox()
    #python_button = driver.find_elements_by_xpath("/html/body/div/div[1]/nav/ul[1]/li[3]/a")
    driver.get (url_salarios)
    driver.find_element_by_id("cmp-salary-search-input").send_keys(busqueda)
    boton = driver.find_element_by_id("cmp-salary-search-submit")
    time.sleep(1)
    boton.click()
    html = driver.page_source
    driver.quit()
    
    soup = BeautifulSoup(html,"html.parser")
    return soup
```

```python
informacion = get_info_page()
```

```python

def salario_empresas(informacion):
    table=informacion.find_all('tr', attrs={'class':'cmp-salary-aggregate-table-entry cmp-sal-separate-row'})
    lst = []


    for e in table:
        dic = {}
        name=e.find('div', attrs={'class':'cmp-sal-description'})
    
        dic["empresa"] = name.find("a").text
        salary=e.find('div', attrs={'class':'cmp-sal-summary'})
        dic["salario"] = salary.find("strong").text
        lst.append(dic)
    return lst
```

```python
busqueda = "consultor-general-Salaries"
empresas = []
for count in range(10,240,10):
    paginas = "?start={}".format(count)
    url_salario = "https://www.indeed.es/salaries/{}{}".format(busqueda,paginas)
    html = requests.get(url_salario).text
    informacion = BeautifulSoup(html,"html.parser")
    empresas += salario_empresas(informacion)
```

```python
sueldo_consultor = pd.DataFrame(empresas)
sueldo_consultor
```
