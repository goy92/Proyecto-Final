```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from pymongo import MongoClient
import nltk
```

```python
cliente = MongoClient('mongodb://localhost:27017/')
db = cliente.linkedin
```

```python
perfiles = db.linkedin.find({"industry": { "$exists": "true"}},{"id":1, "locality":1, "education":1,"skills":1,"industry":1, "experience":1, "specilities":1}).limit(150000)
```

```python
data = pd.DataFrame(perfiles)
```

```python
trade =data[data['industry'] == "International Trade and Development"]
trade_skills =trade['skills']
```

```python
trade_skills
```

```python
trade_skills.fillna("No skills", inplace = True)
```

```python
lista_trade_skills = []
for e in trade_skills:
    if e != "No skills":
        lista_trade_skills += e
print(lista_trade_skills)
```

```python
freq = nltk.FreqDist(lista_trade_skills)
freq
trade_skills = {}
for word, frecuency in freq.most_common(50):
    trade_skills[word]=frecuency
```
